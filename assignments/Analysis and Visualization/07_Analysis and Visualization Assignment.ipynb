## 1. Research Questions
- How do Coaching Styles (positive-reinforcement vs. autonomous-supportive) affect balance performance?
- How does viewpoint (first-person vs. third-person) affect performance and user experience?
- How do Coaching Style and viewpoint interact?

## 2. Hypotheses
- H1: Positive-reinforcement improves performance.
- H2: Positive-reinforcement improves subjective experience.
- H3: Third-person improves performance.
- H4: Third-person improves subjective experience.

## 3. Sample Size
- Determined via G*Power using partial η² = 0.14.
- Final sample size: n = 16.

## 4. Study Design
- 2×2 within-subjects design.
- Factors: Coaching Style × viewpoint.

## 5. Variables
**Independent variables:**
-  Coaching Style: positive-reinforcement, autonomous-supportive
- Viewpoint: first-person, third-person

**Dependent variables:**
- Objective: steadiness, foot height, number of mistakes
- Subjective: social presence, embodiment, enjoyment, difficulty, perception of coach, recommendation, SUS, NASA-TLX

## 6. Statistical Methods
- Two-way repeated-measures ANOVA
- Pairwise t-tests
- Descriptive statistics (mean, SD, SE)

## 7. Descriptive Analysis
- Means, SDs, and SEs reported in tables.
- Boxplots and bar charts visualize results.

## 8. Group Comparisons
- Four conditions (2×2).
- ANOVA appropriate for comparing more than two groups.

## 9. Results Reporting
- ANOVA tables (F, df, p, η²)
- Post-hoc t-tests
- Reports include tables and visualizations

## 10. Visualizations
- Boxplots for steadiness, foot height, mistakes
- Bar charts for questionnaires and NASA-TLX

## 11. Qualitative Analysis
- Comments on tracking accuracy, glasses removal, viewpoint preference, and coach feedback.

## 12. Hypotheses Discussion
- H1: Partially supported (only steadiness improved)
- H2: Supported
- H3: Not supported (first-person steadier)
- H4: Not supported

## 13. Limitations & Future Work
**Limitations:**
- Kinect tracking inaccuracies
- Some participants removed glasses
- Learning effects
- Small sample size
- Tracking affected by clothing/height

**Future Work:**
- Improve tracking
- Add visual/audio/haptic feedback
- Apply system in clinical settings
- Compare VR vs. real-world performance
- ML-based adaptive coaching


```python
# ============================
# Libraries
# ============================
import pandas as pd
import numpy as np
import pingouin as pg
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# ============================
# Data Preparation
# ============================

# Read CSVs
userPerformanceData = pd.read_csv("UserPerformance.csv")
SUS_Data            = pd.read_csv("SUS.csv")
TLX_Data            = pd.read_csv("TLX.csv")
MidQuestion_Data    = pd.read_csv("Mid_Questionnaire.csv")

print(userPerformanceData.describe(include='all'))
print(SUS_Data.describe(include='all'))
print(TLX_Data.describe(include='all'))
print(MidQuestion_Data.describe(include='all'))

# ------ Remove P2 and P12 from UserPerformance.csv
userPerformanceData = userPerformanceData[
    ~userPerformanceData["PID"].isin([2, 12])
]

# ------ SUS: remove P2 and P12
SUS_Data = SUS_Data[~SUS_Data["UserId"].isin([2, 12])]

# ------ TLX: remove P2 and P12
TLX_Data = TLX_Data[~TLX_Data["UserId"].isin([2, 12])]

# ------ Mid Questionnaire: remove P2 and P12 (by PID code)
MidQuestion_Data = MidQuestion_Data[
    ~MidQuestion_Data["PID"].isin(["Me336", "Ka513"])
]

print(MidQuestion_Data.describe(include='all'))

# ============================
# Helper: summary stats
# (similar to rstatix::get_summary_stats type="common")
# ============================

def get_summary_stats(df, cols):
    """
    Returns mean, sd, median, min, max, n for given columns.
    """
    stats = df[cols].agg(['mean', 'std', 'median', 'min', 'max', 'count']).T
    stats.rename(columns={'count': 'n'}, inplace=True)
    return stats.reset_index().rename(columns={'index': 'variable'})

# ============================
# Descriptive Analysis
# ============================

# ---- User performance by CoachingStyle
userPerformance_Descriptive = (
    userPerformanceData
    .groupby("CoachingStyle", as_index=False)
    .apply(lambda g: get_summary_stats(g, ["FootHeight", "Steadiness", "Mistake"]))
)

print("User performance descriptive by CoachingStyle")
print(userPerformance_Descriptive)

# ---- User performance by CoachingStyle and Viewpoint
userPerformance_Descriptive2 = (
    userPerformanceData
    .groupby(["CoachingStyle", "Viewpoint"], as_index=False)
    .apply(lambda g: get_summary_stats(g, ["FootHeight", "Steadiness", "Mistake"]))
)

print("User performance descriptive by CoachingStyle, Viewpoint")
print(userPerformance_Descriptive2)

# ---- SUS
SUS_Descriptive = get_summary_stats(SUS_Data, ["SUS"])
print("SUS Descriptive")
print(SUS_Descriptive)

# ---- TLX by Type
TLX_Descriptive = (
    TLX_Data
    .groupby("Type", as_index=False)
    .apply(lambda g: get_summary_stats(g, ["Score"]))
)
print("TLX Descriptive by Type")
print(TLX_Descriptive)

# ---- MidQuestion by  Coaching Style and Viewpoint
mid_cols = [
    "Embodiment", "Social_Presence", "Enjoyment", "Difficulty",
    "PerceptionCoach", "Recommendation"
]

MidQuestion_Descriptive = (
    MidQuestion_Data
    .groupby([" Coaching Style", "Viewpoint"], as_index=False)
    .apply(lambda g: get_summary_stats(g, mid_cols))
)
print("Mid-questionnaire Descriptive")
print(MidQuestion_Descriptive)

# ============================
# Statistical Analysis
# ============================

# Convert to categorical (factor in R)
userPerformanceData["PID"]          = userPerformanceData["PID"].astype("category")
userPerformanceData["CoachingStyle"] = userPerformanceData["CoachingStyle"].astype("category")
userPerformanceData["Viewpoint"]    = userPerformanceData["Viewpoint"].astype("category")

# ---------- Repeated-measures ANOVA: FootHeight
# R: ezANOVA(..., dv = FootHeight, wid = PID, within = .(Viewpoint, CoachingStyle))

foot_aov = pg.rm_anova(
    data=userPerformanceData,
    dv="FootHeight",
    within=["Viewpoint", "CoachingStyle"],
    subject="PID",
    detailed=True
)
print("ANOVA FootHeight")
print(foot_aov)

# ---------- Steadiness
steadiness_aov = pg.rm_anova(
    data=userPerformanceData,
    dv="Steadiness",
    within=["CoachingStyle", "Viewpoint"],
    subject="PID",
    detailed=True
)
print("ANOVA Steadiness")
print(steadiness_aov)

# ---------- Mistake
mistake_aov = pg.rm_anova(
    data=userPerformanceData,
    dv="Mistake",
    within=["CoachingStyle", "Viewpoint"],
    subject="PID",
    detailed=True
)
print("ANOVA Mistake")
print(mistake_aov)

# ---------- Interpret Cohen's d
def interpret_cohens_d(d):
    if abs(d) < 0.2:
        return "negligible"
    elif abs(d) < 0.5:
        return "small"
    elif abs(d) < 0.8:
        return "medium"
    else:
        return "large"

d_value = 0.0166139261
print("Cohen's d:", d_value, "->", interpret_cohens_d(d_value))

# ============================
# Pairwise t-tests (pingouin)
# ============================

# ---- Steadiness ~ Viewpoint, paired
pair_stead_viewpoint = pg.pairwise_ttests(
    data=userPerformanceData,
    dv="Steadiness",
    within="Viewpoint",
    subject="PID",
    padjust="bonf"
)
print("Pairwise t-test: Steadiness ~ Viewpoint (paired)")
print(pair_stead_viewpoint)

# ---- Steadiness ~ CoachingStyle, paired
pair_stead_style = pg.pairwise_ttests(
    data=userPerformanceData,
    dv="Steadiness",
    within="CoachingStyle",
    subject="PID",
    padjust="bonf"
)
print("Pairwise t-test: Steadiness ~ CoachingStyle (paired)")
print(pair_stead_style)

# Helper for grouped pairwise within tests
def grouped_pairwise_within(df, group_cols, dv, within, subject, padjust="bonf"):
    results = []
    for keys, sub in df.groupby(group_cols):
        res = pg.pairwise_ttests(
            data=sub,
            dv=dv,
            within=within,
            subject=subject,
            padjust=padjust
        )
        # Attach group levels
        if not isinstance(keys, tuple):
            keys = (keys,)
        for col, val in zip(group_cols, keys):
            res[col] = val
        results.append(res)
    return pd.concat(results, ignore_index=True)

# ---- Steadiness ~ CoachingStyle, grouped by Viewpoint
pair_stead_style_by_view = grouped_pairwise_within(
    userPerformanceData,
    group_cols=["Viewpoint"],
    dv="Steadiness",
    within="CoachingStyle",
    subject="PID"
)
print("Steadiness ~ CoachingStyle, grouped by Viewpoint")
print(pair_stead_style_by_view)

# ---- Steadiness ~ Viewpoint, grouped by CoachingStyle
pair_stead_view_by_style = grouped_pairwise_within(
    userPerformanceData,
    group_cols=["CoachingStyle"],
    dv="Steadiness",
    within="Viewpoint",
    subject="PID"
)
print("Steadiness ~ Viewpoint, grouped by CoachingStyle")
print(pair_stead_view_by_style)

# ---- Steadiness ~ CoachingStyle, grouped by Pose & Viewpoint
pair_stead_style_by_pose_view = grouped_pairwise_within(
    userPerformanceData,
    group_cols=["Pose", "Viewpoint"],
    dv="Steadiness",
    within="CoachingStyle",
    subject="PID"
)
print("Steadiness ~ CoachingStyle, grouped by Pose & Viewpoint")
print(pair_stead_style_by_pose_view)

# ---- Steadiness ~ Viewpoint, grouped by Pose & CoachingStyle
pair_stead_view_by_pose_style = grouped_pairwise_within(
    userPerformanceData,
    group_cols=["Pose", "CoachingStyle"],
    dv="Steadiness",
    within="Viewpoint",
    subject="PID"
)
print("Steadiness ~ Viewpoint, grouped by Pose & CoachingStyle")
print(pair_stead_view_by_pose_style)

# ---- FootHeight pairwise (examples; rest analogous to above patterns)
pair_foot_view_by_style = grouped_pairwise_within(
    userPerformanceData,
    group_cols=["CoachingStyle"],
    dv="FootHeight",
    within="Viewpoint",
    subject="PID"
)
print("FootHeight ~ Viewpoint, grouped by CoachingStyle")
print(pair_foot_view_by_style)

pair_foot_view_by_style_pose = grouped_pairwise_within(
    userPerformanceData,
    group_cols=["CoachingStyle", "Pose"],
    dv="FootHeight",
    within="Viewpoint",
    subject="PID"
)
print("FootHeight ~ Viewpoint, grouped by CoachingStyle & Pose")
print(pair_foot_view_by_style_pose)

pair_foot_style_by_view_pose = grouped_pairwise_within(
    userPerformanceData,
    group_cols=["Viewpoint", "Pose"],
    dv="FootHeight",
    within="CoachingStyle",
    subject="PID"
)
print("FootHeight ~ CoachingStyle, grouped by Viewpoint & Pose")
print(pair_foot_style_by_view_pose)

# ---- Mistake pairwise (examples)
pair_mistake_view_by_style_pose = grouped_pairwise_within(
    userPerformanceData,
    group_cols=["CoachingStyle", "Pose"],
    dv="Mistake",
    within="Viewpoint",
    subject="PID"
)
print("Mistake ~ Viewpoint, grouped by CoachingStyle & Pose")
print(pair_mistake_view_by_style_pose)

pair_mistake_style_by_view_pose = grouped_pairwise_within(
    userPerformanceData,
    group_cols=["Viewpoint", "Pose"],
    dv="Mistake",
    within="CoachingStyle",
    subject="PID"
)
print("Mistake ~ CoachingStyle, grouped by Viewpoint & Pose")
print(pair_mistake_style_by_view_pose)

# ---------- Mid-Questionnaire pairwise tests

# Enjoyment ~  Coaching Style (paired)
mid_enjoyment = pg.pairwise_ttests(
    data=MidQuestion_Data,
    dv="Enjoyment",
    within=" Coaching Style",
    subject="PID",  # assuming PID is the subject ID; change if needed
    padjust="bonf"
)
print("Mid Enjoyment ~  Coaching Style")
print(mid_enjoyment)

# Difficulty ~  Coaching Style
mid_difficulty = pg.pairwise_ttests(
    data=MidQuestion_Data,
    dv="Difficulty",
    within=" Coaching Style",
    subject="PID",
    padjust="bonf"
)
print("Mid Difficulty ~  Coaching Style")
print(mid_difficulty)

# Embodiment ~ Viewpoint
mid_embodiment_view = pg.pairwise_ttests(
    data=MidQuestion_Data,
    dv="Embodiment",
    within="Viewpoint",
    subject="PID",
    padjust="bonf"
)
print("Mid Embodiment ~ Viewpoint")
print(mid_embodiment_view)

# Social_Presence ~  Coaching Style, grouped by Viewpoint
mid_social_by_view = grouped_pairwise_within(
    MidQuestion_Data,
    group_cols=["Viewpoint"],
    dv="Social_Presence",
    within=" Coaching Style",
    subject="PID"
)
print("Mid Social_Presence ~  Coaching Style, by Viewpoint")
print(mid_social_by_view)

# PerceptionCoach ~  Coaching Style
mid_perception = pg.pairwise_ttests(
    data=MidQuestion_Data,
    dv="PerceptionCoach",
    within=" Coaching Style",
    subject="PID",
    padjust="bonf"
)
print("Mid PerceptionCoach ~  Coaching Style")
print(mid_perception)

mid_perception_by_view = grouped_pairwise_within(
    MidQuestion_Data,
    group_cols=["Viewpoint"],
    dv="PerceptionCoach",
    within=" Coaching Style",
    subject="PID"
)
print("Mid PerceptionCoach ~  Coaching Style, by Viewpoint")
print(mid_perception_by_view)

# Recommendation ~  Coaching Style
mid_recommend = pg.pairwise_ttests(
    data=MidQuestion_Data,
    dv="Recommendation",
    within=" Coaching Style",
    subject="PID",
    padjust="bonf"
)
print("Mid Recommendation ~  Coaching Style")
print(mid_recommend)

mid_recommend_by_view = grouped_pairwise_within(
    MidQuestion_Data,
    group_cols=["Viewpoint"],
    dv="Recommendation",
    within=" Coaching Style",
    subject="PID"
)
print("Mid Recommendation ~  Coaching Style, by Viewpoint")
print(mid_recommend_by_view)

# ============================
# Visualization
# ============================

# --- Rename columns / levels (mapvalues)
userPerformanceData["CoachingStyle"] = (
    userPerformanceData["CoachingStyle"]
    .replace({
        "Autonomous Style": "Autonomous-Supportive Style",
        "Controlling Style": "Controlled Style"
    })
)

userPerformanceData["CoachingStyleName"] = (
    userPerformanceData["CoachingStyle"]
    .replace({
        "Autonomous-Supportive Style": "Autonomous-SS.",
        "Controlled Style": "Controlled S."
    })
)

# -------- Steadiness boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(
    data=userPerformanceData,
    x="CoachingStyleName",
    y="Steadiness",
    hue="Viewpoint"
)
sns.pointplot(
    data=userPerformanceData,
    x="CoachingStyleName",
    y="Steadiness",
    hue="Viewpoint",
    dodge=0.35,
    join=False,
    markers="o",
    linestyles="",
    errorbar=None
)
plt.xlabel("")
plt.ylabel("Steadiness (m)")
plt.legend(title="Viewpoint", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# -------- FootHeight boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(
    data=userPerformanceData,
    x="CoachingStyleName",
    y="FootHeight",
    hue="Viewpoint"
)
sns.pointplot(
    data=userPerformanceData,
    x="CoachingStyleName",
    y="FootHeight",
    hue="Viewpoint",
    dodge=0.35,
    join=False,
    markers="o",
    linestyles="",
    errorbar=None
)
plt.xlabel("")
plt.ylabel("Foot Height (m)")
plt.legend(title="Viewpoint", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# -------- Mistakes barplot
plt.figure(figsize=(6, 4))
sns.barplot(
    data=userPerformanceData,
    x="CoachingStyleName",
    y="Mistake",
    hue="Viewpoint",
    ci="sd"
)
plt.xlabel("")
plt.ylabel("Number of Mistakes")
plt.legend(title="Viewpoint", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# (Combining plots like ggarrange can be done with matplotlib's subplots)

# ============================
# Pose Visualization – Steadiness
# ============================

userPerformanceData["PoseName"] = (
    userPerformanceData["Pose"]
    .replace({
        "1": "Pose 1", "2": "Pose 2", "3": "Pose 3",
        "4": "Pose 4", "5": "Pose 5", "6": "Pose 6"
    })
)

# Boxplot Steadiness by Pose & Viewpoint
plt.figure(figsize=(10, 4))
sns.boxplot(
    data=userPerformanceData,
    x="PoseName",
    y="Steadiness",
    hue="Viewpoint"
)
sns.pointplot(
    data=userPerformanceData,
    x="PoseName",
    y="Steadiness",
    hue="Viewpoint",
    dodge=0.35,
    join=False,
    markers="o",
    linestyles="",
    errorbar=None
)
plt.xlabel("")
plt.ylabel("Steadiness (m)")
plt.legend(title="Viewpoint", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# ============================
# TLX Visualization
# ============================

TLX_Data["TypeName"] = TLX_Data["Type"].replace({
    "Mental Demand": "Mental D.",
    "Physical Demand": "Physical D.",
    "Temporal Demand": "Temporal D.",
    "Performance": "Performance",
    "Effort": "Effort",
    "Frustration": "Frustration"
})

plt.figure(figsize=(6, 4))
sns.barplot(
    data=TLX_Data,
    x="TypeName",
    y="Score",
    ci="sd"
)
plt.xlabel("")
plt.ylabel("NASA TLX")
plt.tight_layout()
plt.show()

# ============================
# Mid Questionnaire Visualization (Examples)
# ============================

# Embodiment
plt.figure(figsize=(6, 4))
sns.barplot(
    data=MidQuestion_Data,
    x=" Coaching Style",
    y="Embodiment",
    hue="Viewpoint",
    ci="sd"
)
plt.ylim(0, 7)
plt.xlabel("")
plt.ylabel("Embodiment")
plt.legend(title="Viewpoint", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# Social Presence
plt.figure(figsize=(6, 4))
sns.barplot(
    data=MidQuestion_Data,
    x=" Coaching Style",
    y="Social_Presence",
    hue="Viewpoint",
    ci="sd"
)
plt.ylim(0, 7)
plt.xlabel("")
plt.ylabel("Social Presence")
plt.legend(title="Viewpoint", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# Enjoyment
plt.figure(figsize=(6, 4))
sns.barplot(
    data=MidQuestion_Data,
    x=" Coaching Style",
    y="Enjoyment",
    hue="Viewpoint",
    ci="sd"
)
plt.ylim(0, 5)
plt.xlabel("")
plt.ylabel("Enjoyment")
plt.legend(title="Viewpoint", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# Difficulty
plt.figure(figsize=(6, 4))
sns.barplot(
    data=MidQuestion_Data,
    x=" Coaching Style",
    y="Difficulty",
    hue="Viewpoint",
    ci="sd"
)
plt.ylim(0, 5)
plt.xlabel("")
plt.ylabel("Difficulty")
plt.legend(title="Viewpoint", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# Perception of Coach
plt.figure(figsize=(6, 4))
sns.barplot(
    data=MidQuestion_Data,
    x=" Coaching Style",
    y="PerceptionCoach",
    hue="Viewpoint",
    ci="sd"
)
plt.ylim(0, 5)
plt.xlabel("")
plt.ylabel("Perception of Coach")
plt.legend(title="Viewpoint", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# Recommendation
plt.figure(figsize=(6, 4))
sns.barplot(
    data=MidQuestion_Data,
    x=" Coaching Style",
    y="Recommendation",
    hue="Viewpoint",
    ci="sd"
)
plt.ylim(0, 11)
plt.xlabel("")
plt.ylabel("Recommendation")
plt.legend(title="Viewpoint", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
