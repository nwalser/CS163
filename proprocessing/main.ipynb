```python
# GENERAL IMPORTS
import matplotlib.pyplot as plt
import requests, re
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from concurrent.futures import ThreadPoolExecutor, as_completed
import threading
from tqdm.notebook import tqdm

plt.rcParams['image.cmap'] = 'inferno'
```

```python
# DOWNLOAD DATA
archive_url = "https://noaadata.apps.nsidc.org/NOAA/G02135/north/daily/geotiff/"
name_pattern = re.compile(r".*extent_v4.0.tif$")
data_dir = "images"
max_workers = 20
```

```python
print_lock = threading.Lock()

def download_file(url, outdir):
    os.makedirs(outdir, exist_ok=True)
    filename = os.path.join(outdir, os.path.basename(url))
    if not os.path.exists(filename):
        file_data = requests.get(url).content
        with open(filename, "wb") as f:
            f.write(file_data)
        with print_lock:
            print("Downloaded", url)
    else:
        with print_lock:
            print("Skipped", url)


def gather_links(url, pattern):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")
    files, folders = [], []
    for link in soup.find_all("a")[1:]:
        href = link.get("href")
        full_url = urljoin(url, href)
        if href.endswith("/"):
            folders.append(full_url)
        elif pattern.match(href):
            files.append(full_url)
    return files, folders


def download_recursive_parallel(url, outdir, pattern, executor):
    os.makedirs(outdir, exist_ok=True)
    files, folders = gather_links(url, pattern)

    futures = [executor.submit(download_file, f, outdir) for f in files]

    for folder in folders:
        subdir = os.path.join(outdir, folder.strip("/").split("/")[-1])
        download_recursive_parallel(folder, subdir, pattern, executor)

    for f in as_completed(futures):
        f.result()


with ThreadPoolExecutor(max_workers=max_workers) as executor:
    download_recursive_parallel(archive_url, data_dir, name_pattern, executor)
```

```python
import os, re
import numpy as np
from PIL import Image

# count matching files
n_files = sum(1 for root, _, files in os.walk(data_dir)
              for file in files if name_pattern.match(file))

img_shape = (448, 304, 3)
images = np.zeros((n_files, *img_shape), dtype='uint8')
dates = np.zeros(n_files, dtype='datetime64[D]')

idx = 0
for root, _, files in os.walk(data_dir):
    for file in files:
        if name_pattern.match(file):
            path = os.path.join(root, file)

            # load image
            img = np.array(Image.open(path).convert('RGB'), dtype='uint8')
            images[idx] = img

            # extract date
            m = re.search(r'_(\d{8})_', file)
            date_str = m.group(1)
            date_iso = f"{date_str[:4]}-{date_str[4:6]}-{date_str[6:]}"
            dates[idx] = np.datetime64(date_iso, 'D')

            idx += 1
```

```python
# CALCULATE AREA
pixel_size = 25
n_images = images.shape[0]
areas = np.zeros(n_images, dtype='int64')

for i in tqdm(range(n_images)):
    mask = np.all(images[i] == 255, axis=-1)
    areas[i] = np.sum(mask) * pixel_size**2
```

```python
# CALCULATE AGE OF SEA ICE
n_images, H, W, _ = images.shape
counter = np.zeros((H, W), dtype=float)
age_stack = np.zeros((n_images, H, W), dtype=float)

for i in tqdm(range(1, n_images)):
    img = images[i]
    mask = np.all(img == 255, axis=-1)

    delta_days = (dates[i] - dates[i-1]).astype(int)
    counter = np.where(mask, counter + delta_days, 0)
    age_stack[i] = counter
```

```python
# CALCULATE LAST TIME FROZEN DATE
last_white_date = np.zeros(img_shape[:2], dtype='datetime64[D]')

for i in tqdm(range(len(images)), total=len(images)):
    mask = np.all(images[i] == 255, axis=-1)
    last_white_date[mask] = dates[i]
```

```python
years = last_white_date.astype('datetime64[Y]').astype(int) + 1970

plt.figure(figsize=(8,6))
im = plt.imshow(years)
cbar = plt.colorbar(im)
cbar.set_label("Year")
plt.title("Last time frozen")
plt.axis("off")
plt.show()
```

```python
idx = 14000

plt.figure(figsize=(6,6))
plt.imshow(age_stack[idx] / 365.25, interpolation='nearest', vmin=0, vmax=4)
plt.title(f"{dates[idx]}")
plt.axis('off')
plt.colorbar(label='Ice age [years]')
plt.show()
```

```python
day_of_year = (dates - dates.astype('datetime64[Y]')).astype(int) + 1
year = dates.astype('datetime64[Y]').astype(int) + 1970

plt.figure(figsize=(10,5))
scatter = plt.scatter(day_of_year, areas, c=year, cmap='inferno', s=5)
plt.xlabel("Day of Year")
plt.ylabel("Area [km²]")
plt.title("Area vs Day of Year (colored by year)")
plt.xticks(range(0, 366, 30))
plt.colorbar(scatter, label='Year')
plt.tight_layout()
plt.show()
```
