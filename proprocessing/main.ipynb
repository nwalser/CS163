```python
# GENERAL IMPORTS
from IPython.display import display
import pandas as pd
from PIL import Image
import numpy as np
import os
import requests, re
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from concurrent.futures import ThreadPoolExecutor, as_completed
import threading
```

```python
# DOWNLOAD DATA
archive_url = "https://noaadata.apps.nsidc.org/NOAA/G02135/north/daily/geotiff/"
name_pattern = re.compile(r".*extent_v4.0.tif$")
data_dir = "data"
max_workers = 20

print_lock = threading.Lock()


def download_file(url, outdir):
    os.makedirs(outdir, exist_ok=True)
    filename = os.path.join(outdir, os.path.basename(url))
    if not os.path.exists(filename):
        file_data = requests.get(url).content
        with open(filename, "wb") as f:
            f.write(file_data)
        with print_lock:
            print("Downloaded", url)
    else:
        with print_lock:
            print("Skipped", url)


def gather_links(url, pattern):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")
    files, folders = [], []
    for link in soup.find_all("a")[1:]:
        href = link.get("href")
        full_url = urljoin(url, href)
        if href.endswith("/"):
            folders.append(full_url)
        elif pattern.match(href):
            files.append(full_url)
    return files, folders


def download_recursive_parallel(url, outdir, pattern, executor):
    os.makedirs(outdir, exist_ok=True)
    files, folders = gather_links(url, pattern)

    futures = [executor.submit(download_file, f, outdir) for f in files]

    for folder in folders:
        subdir = os.path.join(outdir, folder.strip("/").split("/")[-1])
        download_recursive_parallel(folder, subdir, pattern, executor)

    for f in as_completed(futures):
        f.result()


with ThreadPoolExecutor(max_workers=max_workers) as executor:
    download_recursive_parallel(archive_url, data_dir, name_pattern, executor)
```

```python
# LOAD DATA
rows = []

for root, _, files in os.walk(data_dir):
    for file in files:
        if pattern.match(file):
            path = os.path.join(root, file)
            rows.append({
                "filename": file,
                "local_path": path,
                "size_bytes": os.path.getsize(path)
            })

df = pd.DataFrame(rows)
```

```python
# EXTRACT DATE
df["date"] = df["filename"].str.extract(r"_(\d{8})_")[0]
df["date"] = pd.to_datetime(df["date"], format="%Y%m%d")
```

```python
# CALCULATE AREA
pixel_size = 25

def calculate_area(path):
    img = Image.open(path).convert("L")
    arr = np.array(img)
    return np.sum(arr == 255) * pixel_size**2

df["area"] = df["local_path"].apply(calculate_area)
```

```python
# CALCULATE AGE OF SEA ICE
first_image_path = df["local_path"][0]
first_image = Image.open(first_image_path)
img_shape = np.array(first_image).shape[:2]
counter = np.zeros(img_shape, dtype=float)  # use float for fractional days

previous_date = None

def calculate_age(row):
    global counter, previous_date
    img = Image.open(row["local_path"]).convert("RGB")
    arr = np.array(img)
    mask = np.all(arr == 255, axis=-1)

    if previous_date is None:
        delta_days = 0
    else:
        delta_days = (row["date"] - previous_date).days
    previous_date = row["date"]

    counter = np.where(mask, counter + delta_days, 0)
    return counter.copy()

df["age"] = df.apply(calculate_age, axis=1)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df['day_of_year'] = df['date'].dt.dayofyear
df['year'] = df['date'].dt.year

plt.figure(figsize=(10,5))
scatter = plt.scatter(df['day_of_year'], df['area'], c=df['year'], cmap='inferno', s=5)
plt.xlabel("Day of Year")
plt.ylabel("Area [km²]")
plt.title("Area vs Day of Year (colored by year)")
plt.xticks(range(0, 366, 30))
plt.colorbar(scatter, label='Year')
plt.tight_layout()
plt.show()
```
