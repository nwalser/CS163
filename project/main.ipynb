```python
# GENERAL IMPORTS
import matplotlib.pyplot as plt
import requests, re
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from tqdm import tqdm

plt.rcParams['image.cmap'] = 'inferno'
```

```python
# DOWNLOAD DATA
archive_url = "https://noaadata.apps.nsidc.org/NOAA/G02135/north/daily/geotiff/"
name_pattern = re.compile(r".*extent_v4.0.tif$")
data_dir = "images"
max_workers = 20

analysis_start = 1980
analysis_end = 2022
```

```python
def download_recursive(url, outdir, pattern, pbar):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    for link in soup.find_all("a")[1:]:
        href = link.get("href")
        full_url = urljoin(url, href)

        if href.endswith("/"):
            subdir = os.path.join(outdir, os.path.basename(href.strip("/")))
            download_recursive(full_url, subdir, pattern, pbar)
        elif pattern.match(href):
            os.makedirs(outdir, exist_ok=True)
            filename = os.path.join(outdir, os.path.basename(full_url))
            if not os.path.exists(filename):
                with open(filename, "wb") as f:
                    f.write(requests.get(full_url).content)
            pbar.update(1)

with tqdm(desc="Downloading files", unit="file", total=15000) as pbar:
    #download_recursive(archive_url, data_dir, name_pattern, pbar) # uncomment if you want to download the data
    pass
```

```python
import os, re
import numpy as np
from PIL import Image

# count matching files
n_files = sum(1 for root, _, files in os.walk(data_dir)
              for file in files if name_pattern.match(file))

img_shape = (448, 304, 3)
images = np.zeros((n_files, *img_shape), dtype='uint8')
dates = np.zeros(n_files, dtype='datetime64[D]')

idx = 0
for root, _, files in tqdm(os.walk(data_dir)):
    for file in files:
        if name_pattern.match(file):
            path = os.path.join(root, file)

            # load image
            img = np.array(Image.open(path).convert('RGB'), dtype='uint8')
            images[idx] = img

            # extract date
            m = re.search(r'_(\d{8})_', file)
            date_str = m.group(1)
            date_iso = f"{date_str[:4]}-{date_str[4:6]}-{date_str[6:]}"
            dates[idx] = np.datetime64(date_iso, 'D')

            idx += 1
```

```python
# EXTRACT LAND MASSES
landmasses = np.all(images[0] == 119, axis=-1) | np.all(images[0] == 0, axis=-1)
```

```python
# CALCULATE AREA
pixel_size = 25
n_images = images.shape[0]
areas = np.zeros(n_images, dtype='int64')

for i in tqdm(range(n_images)):
    mask = np.all(images[i] == 255, axis=-1)
    areas[i] = np.sum(mask) * pixel_size**2
```

```python
# CALCULATE AGE OF SEA ICE
n_images, H, W, _ = images.shape
counter = np.zeros((H, W), dtype=float)
age_stack = np.zeros((n_images, H, W), dtype=float)

for i in tqdm(range(1, n_images)):
    img = images[i]
    mask = np.all(img == 255, axis=-1)

    delta_days = (dates[i] - dates[i-1]).astype(int)
    counter = np.where(mask, counter + delta_days, 0)
    age_stack[i] = counter
```

```python
# CALCULATE LAST TIME FROZEN DATE
last_white_date = np.full(img_shape[:2], np.datetime64('NaT', 'D'))

for i in tqdm(range(len(images)), total=len(images)):
    mask = np.all(images[i] == 255, axis=-1)
    last_white_date[mask] = dates[i]
```

```python
# CALCULATE FIRST TIME MOLTEN DATE
first_molten_date = np.full(img_shape[:2], np.datetime64('NaT', 'D'))

for i in tqdm(range(len(images))):
    is_molten = ~np.all(images[i] == 255, axis=-1)
    mask = np.isnat(first_molten_date) & is_molten
    first_molten_date[mask] = dates[i]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

years = dates.astype('datetime64[Y]').astype(int) + 1970
unique_years = np.unique(years)

avg_daily_fluct = np.zeros_like(unique_years, dtype=float)
avg_daily_fluct_pct = np.zeros_like(unique_years, dtype=float)

for i, year in enumerate(unique_years):
    mask = years == year
    year_areas = areas[mask]
    year_dates = dates[mask]

    order = np.argsort(year_dates)
    year_areas = year_areas[order]
    year_dates = year_dates[order]

    dA = np.diff(year_areas)
    dt = np.diff(year_dates).astype('timedelta64[D]').astype(int)

    valid = dt > 0
    if np.any(valid):
        daily_changes = np.abs(dA[valid] / dt[valid])
        avg_daily_fluct[i] = np.mean(daily_changes)

        avg_daily_fluct_pct[i] = (avg_daily_fluct[i] / np.mean(year_areas)) * 100
    else:
        avg_daily_fluct[i] = np.nan
        avg_daily_fluct_pct[i] = np.nan


# scale factor for proportional axes
with np.errstate(invalid='ignore', divide='ignore'):
    k = np.nanmean(avg_daily_fluct_pct / avg_daily_fluct)

if not np.isfinite(k) or k == 0:
    k = 1.0


plt.figure(figsize=(12, 6))
ax1 = plt.gca()

# Colors
color_abs = "tab:blue"
color_pct = "tab:red"

# absolute line
line_abs, = ax1.plot(unique_years, avg_daily_fluct,
                     marker='x', linestyle='--', color=color_abs, label="Absolute (km²/day)")

# percent series converted into primary axis units
pct_primary = avg_daily_fluct_pct / k
line_pct, = ax1.plot(unique_years, pct_primary,
                     marker='o', linestyle='-', color=color_pct, label="Percent (%)")

# axis 1 formatting
ax1.set_ylabel("Average Daily Fluctuation", color=color_abs)
ax1.tick_params(axis='y', colors=color_abs)
ax1.yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f"{x:,.0f} km²"))

# proportional secondary axis
ax2 = ax1.secondary_yaxis("right", functions=(lambda x: k*x, lambda x: x/k))
ax2.set_ylabel("Average Daily Fluctuation", color=color_pct)
ax2.tick_params(axis='y', colors=color_pct)
ax2.yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f"{x:.2f}%"))

plt.xticks(unique_years, rotation=90)
plt.xlabel("Year")
plt.title("Average Daily Sea Ice Area Fluctuation")
plt.grid(True, axis='x')
plt.tight_layout()
plt.xlim(analysis_start, analysis_end)

plt.show()
```

```python
import pandas as pd
from matplotlib.ticker import FuncFormatter

df = pd.DataFrame({
    "date": pd.to_datetime(dates.astype(str)),
    "area": areas
})

df["year"] = df["date"].dt.year
df["doy"]  = df["date"].dt.dayofyear

heatmap = df.pivot_table(index="doy", columns="year", values="area")

full_doy = np.arange(1, 367)
heatmap = heatmap.reindex(full_doy)
heatmap = heatmap.interpolate(axis=0, limit_direction='both')

# restrict years
if (analysis_start is not None) or (analysis_end is not None):
    years = np.array(heatmap.columns)
    mask = np.ones_like(years, dtype=bool)
    if analysis_start is not None:
        mask &= years >= analysis_start
    if analysis_end is not None:
        mask &= years <= analysis_end
    heatmap = heatmap.loc[:, years[mask]]

years = list(heatmap.columns)
days  = list(heatmap.index)

plt.figure(figsize=(10, 6))
im = plt.imshow(heatmap.values, aspect='auto', interpolation='nearest', origin='lower', vmin=0)

cbar = plt.colorbar(im, label="Sea Ice Extent")
cbar.ax.yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f"{x:,.0f} km²"))

plt.ylabel("Day of Year")
plt.xlabel('Year')
plt.title("Daily Arctic Sea Ice Extent")

ax = plt.gca()

x_max, y_max = [], []
x_min, y_min = [], []

for col_idx, year in enumerate(years):
    col = heatmap[year]

    max_doy = col.idxmax()
    min_doy = col.idxmin()

    row_idx_max = days.index(max_doy)
    row_idx_min = days.index(min_doy)

    x_max.append(col_idx)
    y_max.append(row_idx_max)

    x_min.append(col_idx)
    y_min.append(row_idx_min)

ax.scatter(x_max, y_max, s=40, facecolors='black', label="Maximal Extent")
ax.scatter(x_min, y_min, s=40, facecolors='white', label="Minimal Extent")

ax.set_xticks(np.arange(len(years)))
ax.set_xticklabels(years, rotation=90)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import os
from pathlib import Path
import numpy as np
from PIL import Image
from tqdm import tqdm


def sea_ice_rgba(img: np.ndarray) -> np.ndarray:
    """
    Convert RGB → RGBA where:
    - pixels exactly [255,255,255] stay white & opaque
    - everything else becomes transparent.
    """
    mask = np.all(img == 255, axis=-1)

    h, w, _ = img.shape
    rgba = np.zeros((h, w, 4), dtype=np.uint8)

    rgba[mask, :3] = 255
    rgba[mask, 3] = 255

    return rgba


def save_daily_sea_ice_pngs(images: np.ndarray,
                            dates: np.ndarray,
                            out_dir: str,
                            start_year: int = None,
                            end_year: int = None):
    """
    Save daily sea ice PNGs, optionally filtering by year range.

    Parameters:
        images      - array of RGB images
        dates       - array of datetime64[D]
        out_dir     - output directory
        start_year  - first year to include (optional)
        end_year    - last year to include (optional)
    """

    out_path = Path(out_dir)
    out_path.mkdir(parents=True, exist_ok=True)

    # Map date → last (or only) image for that date
    date_to_img = {}
    for img, d in zip(images, dates):
        date_to_img[d] = img  # last wins

    start_date = dates.min()
    end_date = dates.max()

    all_days = np.arange(
        start_date, end_date + np.timedelta64(1, "D"), dtype="datetime64[D]"
    )

    last_img = None

    for day in tqdm(all_days):

        # --- Filter by year range -------------------------
        year = int(str(day)[:4])
        if start_year is not None and year < start_year:
            continue
        if end_year is not None and year > end_year:
            continue
        # ---------------------------------------------------

        if day in date_to_img:
            last_img = date_to_img[day]

        if last_img is None:
            continue  # before first available image

        # Format components
        day_str = np.datetime_as_string(day, unit="D")
        yyyy = day_str[:4]
        mm = day_str[5:7]
        yyyymmdd = day_str.replace("-", "")

        # Create year/month subfolder
        subdir = out_path / yyyy / mm
        subdir.mkdir(parents=True, exist_ok=True)

        fname = subdir / f"{yyyymmdd}.png"

        # Skip if exists
        if fname.exists():
            continue

        rgba = sea_ice_rgba(last_img)
        Image.fromarray(rgba).save(fname)
```

```python
save_daily_sea_ice_pngs(
    images,
    dates,
    out_dir="daily_sea_ice",
    start_year=analysis_start,
    end_year=analysis_end
)
```

```python
from typing import Tuple, Literal

import numpy as np
from PIL import Image
from pyproj import Transformer


ProjectionType = Literal["north", "south"]


def build_nsidc_overlay_texture(
    src: str | Image.Image,
    extent_meters: Tuple[float, float, float, float],
    projection: ProjectionType,
    out_width: int = 2048,
) -> Image.Image:
    """
    Reproject a polar stereographic image into an equirectangular RGBA image,
    mimicking the logic of useNSIDCOverlayTexture in TypeScript.

    Parameters
    ----------
    src : str | PIL.Image.Image
        Path to the source image or an already-opened Pillow Image.
        The image is assumed to already be in the NSIDC polar stereographic projection
        with the given `extent_meters`.
    extent_meters : (xmin, ymin, xmax, ymax)
        Bounding box of the source image in meters in the polar stereographic CRS.
    projection : "north" | "south"
        Which polar projection to use (NSIDC standard EPSG codes 3413/3031).
    out_width : int, optional
        Width of the output equirectangular image in pixels (default 2048).
        Height will be out_width * 0.5 (for 360x180 degrees).

    Returns
    -------
    PIL.Image.Image
        The reprojected equirectangular RGBA image.
    """

    # ---- Load image ----
    if isinstance(src, str):
        img = Image.open(src).convert("RGBA")
    else:
        img = src.convert("RGBA")

    src_arr = np.array(img)  # shape (H, W, 4)
    src_h, src_w, _ = src_arr.shape

    # ---- Target canvas (equirectangular) ----
    aspect = 0.5  # 360x180 deg → H = W/2
    dst_w = int(out_width)
    dst_h = int(round(out_width * aspect))

    xmin, ymin, xmax, ymax = extent_meters
    dx = xmax - xmin
    dy = ymax - ymin

    if dx == 0 or dy == 0:
        raise ValueError("Invalid extent_meters: dx or dy is zero.")

    inv_dx = 1.0 / dx
    inv_dy = 1.0 / dy

    # ---- Precompute lon/lat for each column/row (same formulas as TS) ----
    # lon(x) = -180 + (x / (W-1)) * 360
    lon_per_x = np.linspace(-180.0, 180.0, dst_w, endpoint=True)
    # lat(y) =  90 - (y / (H-1)) * 180
    lat_per_y = np.linspace(90.0, -90.0, dst_h, endpoint=True)

    lon_grid, lat_grid = np.meshgrid(lon_per_x, lat_per_y)  # shape (H, W)

    # ---- 4326 → polar stereographic ----
    if projection == "north":
        polar_crs = "EPSG:3413"  # NSIDC Sea Ice Polar Stereographic North
    else:
        polar_crs = "EPSG:3031"  # NSIDC Sea Ice Polar Stereographic South

    transformer = Transformer.from_crs("EPSG:4326", polar_crs, always_xy=True)
    xm, ym = transformer.transform(lon_grid, lat_grid)  # shape (H, W)

    # ---- Normalize into [0, 1] across the provided image extent ----
    u = (xm - xmin) * inv_dx  # shape (H, W)
    v = (ym - ymin) * inv_dy  # shape (H, W)

    # Mask of pixels that fall inside the source raster footprint
    inside = (u >= 0.0) & (u <= 1.0) & (v >= 0.0) & (v <= 1.0)

    # ---- Allocate destination buffer (RGBA) ----
    dst = np.zeros((dst_h, dst_w, 4), dtype=np.uint8)

    # ---- Bilinear resampling for inside pixels ----
    if np.any(inside):
        # Flatten mask indices
        u_i = u[inside]
        v_i = v[inside]

        # Map normalized coords → source pixel coords
        sx = u_i * (src_w - 1)
        sy = (1.0 - v_i) * (src_h - 1)  # y-down in image

        x0 = np.floor(sx).astype(np.int32)
        y0 = np.floor(sy).astype(np.int32)
        x1 = np.clip(x0 + 1, 0, src_w - 1)
        y1 = np.clip(y0 + 1, 0, src_h - 1)

        tx = sx - x0
        ty = sy - y0

        # Gather surrounding 4 pixels (RGBA) as float for interpolation
        c00 = src_arr[y0, x0].astype(np.float32)  # top-left
        c10 = src_arr[y0, x1].astype(np.float32)  # top-right
        c01 = src_arr[y1, x0].astype(np.float32)  # bottom-left
        c11 = src_arr[y1, x1].astype(np.float32)  # bottom-right

        # Bilinear interpolation, channel-wise
        # v0 = c00 + (c10 - c00) * tx
        # v1 = c01 + (c11 - c01) * tx
        # c  = v0 + (v1 - v0) * ty
        tx = tx[:, None]  # (N, 1) for broadcasting over channels
        ty = ty[:, None]

        v0 = c00 + (c10 - c00) * tx
        v1 = c01 + (c11 - c01) * tx
        c = v0 + (v1 - v0) * ty  # (N, 4)

        # Clamp and cast back to uint8
        c = np.clip(c, 0, 255).astype(np.uint8)

        # Write back into destination
        dst[inside] = c

    # Return as Pillow image
    dst_img = Image.fromarray(dst)
    return dst_img
```

```python
import os
from pathlib import Path
from PIL import Image

extent_meters = (-3850000.0, -5350000.0, 3750000.0, 5850000.0)

SRC_ROOT = Path("./daily_sea_ice")
DST_ROOT = Path("./daily_sea_ice_textures")

all_pngs = list(SRC_ROOT.rglob("*.png"))

for src_file in tqdm(all_pngs, desc="Reprojecting", unit="file"):
    rel = src_file.relative_to(SRC_ROOT)
    dst_file = DST_ROOT / rel

    if dst_file.exists():
        continue

    dst_file.parent.mkdir(parents=True, exist_ok=True)

    out_img = build_nsidc_overlay_texture(
        src=str(src_file),
        extent_meters=extent_meters,
        projection="north",
        out_width=512,
    )

    out_img.save(dst_file)
```
