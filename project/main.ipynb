```python
# GENERAL IMPORTS
import matplotlib.pyplot as plt
import requests, re
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin
from tqdm import tqdm

plt.rcParams['image.cmap'] = 'inferno'
```

```python
# DOWNLOAD DATA
archive_url = "https://noaadata.apps.nsidc.org/NOAA/G02135/north/daily/geotiff/"
name_pattern = re.compile(r".*extent_v4.0.tif$")
data_dir = "images"
max_workers = 20
```

```python
def download_recursive(url, outdir, pattern, pbar):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    for link in soup.find_all("a")[1:]:
        href = link.get("href")
        full_url = urljoin(url, href)

        if href.endswith("/"):
            subdir = os.path.join(outdir, os.path.basename(href.strip("/")))
            download_recursive(full_url, subdir, pattern, pbar)
        elif pattern.match(href):
            os.makedirs(outdir, exist_ok=True)
            filename = os.path.join(outdir, os.path.basename(full_url))
            if not os.path.exists(filename):
                with open(filename, "wb") as f:
                    f.write(requests.get(full_url).content)
            pbar.update(1)

with tqdm(desc="Downloading files", unit="file", total=15000) as pbar:
    #download_recursive(archive_url, data_dir, name_pattern, pbar) # uncomment if you want to download the data
    pass
```

```python
import os, re
import numpy as np
from PIL import Image

# count matching files
n_files = sum(1 for root, _, files in os.walk(data_dir)
              for file in files if name_pattern.match(file))

img_shape = (448, 304, 3)
images = np.zeros((n_files, *img_shape), dtype='uint8')
dates = np.zeros(n_files, dtype='datetime64[D]')

idx = 0
for root, _, files in tqdm(os.walk(data_dir)):
    for file in files:
        if name_pattern.match(file):
            path = os.path.join(root, file)

            # load image
            img = np.array(Image.open(path).convert('RGB'), dtype='uint8')
            images[idx] = img

            # extract date
            m = re.search(r'_(\d{8})_', file)
            date_str = m.group(1)
            date_iso = f"{date_str[:4]}-{date_str[4:6]}-{date_str[6:]}"
            dates[idx] = np.datetime64(date_iso, 'D')

            idx += 1
```

```python
# EXTRACT LAND MASSES
landmasses = np.all(images[0] == 119, axis=-1) | np.all(images[0] == 0, axis=-1)
```

```python
# CALCULATE AREA
pixel_size = 25
n_images = images.shape[0]
areas = np.zeros(n_images, dtype='int64')

for i in tqdm(range(n_images)):
    mask = np.all(images[i] == 255, axis=-1)
    areas[i] = np.sum(mask) * pixel_size**2
```

```python
# CALCULATE AGE OF SEA ICE
n_images, H, W, _ = images.shape
counter = np.zeros((H, W), dtype=float)
age_stack = np.zeros((n_images, H, W), dtype=float)

for i in tqdm(range(1, n_images)):
    img = images[i]
    mask = np.all(img == 255, axis=-1)

    delta_days = (dates[i] - dates[i-1]).astype(int)
    counter = np.where(mask, counter + delta_days, 0)
    age_stack[i] = counter
```

```python
# CALCULATE LAST TIME FROZEN DATE
last_white_date = np.full(img_shape[:2], np.datetime64('NaT', 'D'))

for i in tqdm(range(len(images)), total=len(images)):
    mask = np.all(images[i] == 255, axis=-1)
    last_white_date[mask] = dates[i]
```

```python
# CALCULATE FIRST TIME MOLTEN DATE
first_molten_date = np.full(img_shape[:2], np.datetime64('NaT', 'D'))

for i in tqdm(range(len(images))):
    is_molten = ~np.all(images[i] == 255, axis=-1)
    mask = np.isnat(first_molten_date) & is_molten
    first_molten_date[mask] = dates[i]
```

```python
years = last_white_date.astype('datetime64[Y]').astype(int) + 1970

plt.figure(figsize=(8, 6))
im = plt.imshow(years, cmap='inferno', vmin=1980, vmax=2025)
cbar = plt.colorbar(im)
cbar.set_label("Year")
plt.title("Last Time Frozen")
plt.axis("off")

plt.contourf(landmasses, levels=[0.5, 1], colors="gray", zorder=2)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Convert last_white_date → years (float32) with NaN for NaT ---
# years_int will be int64, which some backends dislike for imshow; we'll cast to float32.
years_int = last_white_date.astype('datetime64[Y]').astype('int64') + 1970
years_f32 = years_int.astype('float32')

# Put NaN where dates were NaT
years_f32[np.isnat(last_white_date)] = np.nan

# --- Mask: transparent where <1980 or NaN ---
mask = (years_f32 < 1980) | np.isnan(years_f32)
years_masked = np.ma.array(years_f32, mask=mask)

# --- Colormap with transparent for masked ("bad") values ---
cmap = plt.get_cmap('inferno').copy()
cmap.set_bad(alpha=0)  # masked/NaN pixels become fully transparent

# --- Landmasses to float32 for contourf ---
land_f32 = landmasses.astype('float32')

# --- Plot (no colorbar) ---
plt.figure(figsize=(8, 6))
plt.imshow(years_masked, cmap=cmap, vmin=1980, vmax=2025)

# Gray landmasses on top
plt.contourf(land_f32, levels=[0.5, 1], colors="gray", zorder=2)

# Remove all borders and axes
plt.axis("off")
plt.margins(0, 0)
plt.gca().set_position((0.0, 0.0, 1.0, 1.0))

# --- Save transparent PNG ---
plt.savefig("last_time_frozen.png", dpi=300, transparent=True, bbox_inches=None, pad_inches=0)
plt.show()

```

```python
day_of_year = (dates - dates.astype('datetime64[Y]')).astype(int) + 1
year = dates.astype('datetime64[Y]').astype(int) + 1970

plt.figure(figsize=(10,5))
scatter = plt.scatter(day_of_year, areas, c=year, cmap='inferno', s=5)
plt.xlabel("Day of Year")
plt.ylabel("Area [km²]")
plt.title("Area vs Day of Year (colored by year)")
plt.xticks(range(0, 366, 30))
plt.colorbar(scatter, label='Year')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
years = dates.astype('datetime64[Y]').astype(int) + 1970

unique_years = np.unique(years)
max_doy = np.zeros(len(unique_years))
min_doy = np.zeros(len(unique_years))

for i, year in enumerate(unique_years):
    mask = years == year
    year_areas = areas[mask]
    year_dates = dates[mask]

    day_of_year = (year_dates - year_dates.astype('datetime64[Y]')).astype(int) + 1

    max_doy[i] = day_of_year[np.argmax(year_areas)]
    min_doy[i] = day_of_year[np.argmin(year_areas)]

plt.figure(figsize=(10,5))
plt.plot(unique_years, max_doy, label='Day of Maximum Extent', marker='o')
plt.plot(unique_years, min_doy, label='Day of Minimum Extent', marker='o')

plt.title("Day of Year for Maximum and Minimum Sea Ice Extent")
plt.xlabel("Year")
plt.ylabel("Day of Year (1–365)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
years = dates.astype('datetime64[Y]').astype(int) + 1970

unique_years = np.unique(years)
avg_area = np.zeros_like(unique_years, dtype=float)
min_area = np.zeros_like(unique_years, dtype=float)
max_area = np.zeros_like(unique_years, dtype=float)
never_melted_area = np.zeros_like(unique_years, dtype=float)

for i, year in enumerate(unique_years):
    mask = years == year
    avg_area[i] = np.mean(areas[mask])
    min_area[i] = np.min(areas[mask])
    max_area[i] = np.max(areas[mask])

    cutoff = np.datetime64(f'{year}-12-31', 'D')
    still_never = (~landmasses) & (np.isnat(first_molten_date) | (first_molten_date > cutoff))
    never_melted_area[i] = np.sum(still_never) * (pixel_size ** 2)


plt.figure(figsize=(10, 5))
plt.plot(unique_years, avg_area, marker='o', label='Mean area')
plt.fill_between(unique_years, min_area, max_area, alpha=0.5, label='Minimum and Maximum range')
plt.plot(unique_years, never_melted_area, marker='o', label="Never Melted area")

plt.xlabel('Year')
plt.ylabel('Area (km²)')
plt.title('Average, Minimum and Maximum Area per Year')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
